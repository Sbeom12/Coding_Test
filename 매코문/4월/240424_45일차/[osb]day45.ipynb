```python
def solution(numer1, denom1, numer2, denom2):
    numerator = numer1 * denom2 + numer2 * denom1
    denominator = denom1*denom2
    min_val =  min(denominator, numerator)
    i = 2
    while True:
        print(i, numerator, denominator)
        if denominator % i == 0 and numerator % i == 0:
            numerator = numerator // i
            denominator = denominator // i
            min_val =  min(denominator, numerator)
            i = 1
        i += 1
        if i > min_val:
            break
    return [numerator, denominator]
```

# 최소값 만들기
![alt text](image-1.png)

```python
def solution(A,B):
    answer = 0
    A = sorted(A)
    B = sorted(B, reverse=True)
    for num1, num2 in zip(A, B):
        answer += (num1*num2)
    return answer
```

# 주사위 게임
![alt text](image-2.png)

```python
def solution(a, b, c, d):
    answer = 0
    num_dict= {}
    for num in [a,b,c,d]:
        if num in num_dict:
            num_dict[num] += 1
        else:
            num_dict[num] = 1
    dict_len = len(num_dict)
    
    if dict_len == 1:
        return 1111 * a
    elif dict_len == 4:
        return min(a,b,c,d)
    elif dict_len == 3:
        temp = 1
        for key, value in num_dict.items():
            if value == 1:
                temp *= key
        return temp
    else: # dict_len == 2
        num_list = list(num_dict.keys())
        num1, num2 = num_list[0], num_list[1] 
        if list(num_dict.values())[0] == 2: # 2, 2인경우
            return (num1 + num2) * abs(num1 - num2)
        else:
            if list(num_dict.values())[0] == 3: # 3,1인경우
                return (num1*10 + num2)**2
            else:
                return (num2*10 + num1)**2
```

# 피보나치 수
![alt text](image-3.png)

```python
def solution(n):
    fibo_0, fibo_1 = 0, 1
    answer = 0
    # DP로 풀기
    for i in range(n-1):
        answer = fibo_0+fibo_1
        fibo_0 = fibo_1
        fibo_1 = answer
    return answer%1234567
```

# 치킨 쿠폰
![alt text](image-4.png)

```python
def solution(chicken):
    answer = 0
    rest = 0
    while chicken >= 1:
        rest += chicken%10
        chicken = chicken//10
        answer += chicken
        if rest >= 10:
            answer += 1
            rest -=9
    return answer
```
