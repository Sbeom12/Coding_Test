# 예제

```python
try_num = int(input())
for _ in range(try_num):
    ox_data = input().upper()
    answer = 0
    temp = 0
    for data in ox_data:
        if data == 'O':
            temp += 1
            answer += temp
        elif data == 'X':
            temp = 0
    print(answer)
```

# split(), split(" ")의 차이
---
* split()은 공백이 1개이건 2개이건 상관없이 1개로 보고 처리
    * \t, \n 탭과 엔터도 처리해줌.
* split(" ") 공백 1개를 각각의 공백으로 따로 처리.

```python
string = "word1 word2  word3    word4     "
print(string.split())
print(string.split(" "))
```
