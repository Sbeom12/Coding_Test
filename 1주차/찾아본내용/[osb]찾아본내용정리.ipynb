```python
try_num = int(input())
for _ in range(try_num):
    ox_data = input().upper()
    answer = 0
    temp = 0
    for data in ox_data:
        if data == 'O':
            temp += 1
            answer += temp
        elif data == 'X':
            temp = 0
    print(answer)
```
