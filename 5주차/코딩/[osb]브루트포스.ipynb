# 블랙잭



## 풀이
---
* 시간 : 1초
* 메모리 : 128MB
* 완전탐색 알고리즘 사용.
* 서로다른 N장 중에 3장을 선택하는 문제.
    * itertools. 사용
    * 모든 조합을 구한뒤 M 만큼 뺴준 값을 절대값을 취해서 처리.

```python
from itertools import combinations
N, M = map(int, input().split())
numlist = list(map(int, input().split()))
temp = list(combinations(numlist, 3))
blackjack = []
for number in temp:
    sum_number = number[0] + number[1] + number[2]
    if sum_number <= M:
        blackjack.append(sum_number)
blackjack.sort(key=lambda x: abs(x-M))
print(blackjack[0])
```

```python
# Another way
# 입력 받기
N, M = map(int, input().split(" "))
Card = list(map(int, input().split()))

# 카드 3가지를 뽑는 모든 경우의 수를 찾음.
sum = []
for c1 in range(len(Card)):
  if c1 > len(Card) - 3:
    break
  card1 = Card[c1]
  for c2 in range(c1+1, len(Card)-1):
      card2 = Card[c2]
      for c3 in range(c2+1, len(Card)):
          card3 = Card[c3]
          sum.append(card1 + card2 + card3)
# M에 가장 가까우면서 넘지않는 카드 출력
print(max([i for i in sum if i <= M ]))
```

## 검산
--
* N = 100, M = 300,000
* O(NlonN), Memory = 100* 4byte * 2 = 800byte

# 분해합.


## 풀이
* 제한시간: 2초
* 메모리 : 메모리 제한
* 낮은 수 부터 해서 진행

```python
num = int(input())
answer = 0
for i in range(1, num+1):
    str_num = str(i)
    sum = 0
    for number in str_num:
        sum += int(number)
    if i + sum ==num:
        if i == num:
            answer = 0
            break
        else:
            answer = i
            break
print(answer)
```

```python
num = int(input())
# 자리수 파악
# 자연수 0부터 본인까지 전부 탐지 및 자리 수 더하기
result = 0;
for i in range(1, num+1):
  temp = i + sum(map(int, str(i)))
  if temp == num: # 최소값 출력을 위해 사용.
    result = i
    break
print(result)
```

# 체스판 다시칠하기



## 풀이
---
* 시간 2초
* 메모리 128MB
* 완전 탐색.

```python
N, M= map(int, input().split())
# 판 기본판 생성 8 x 8
Black_board = [['B' if (x+y)%2 == 0 else 'W' for y in range(N)] for x in range(8)]
White_board = [['W' if (x+y)%2 == 0 else 'B' for y in range(N)] for x in range(8)]
# 판 입력 받기
chess_list = [input() for _ in range(N)]
answer= []

for n in range(N-7):
    for m in range(M-7):
        temp = 0
        for y, line in enumerate(Black_board):
            for x, chess in enumerate(line):
                if chess != chess_list[y+n][x+m]:
                    temp += 1
        answer.append(temp)
        temp = 0
        for y, line in enumerate(White_board):
            for x, chess in enumerate(line):
                if chess != chess_list[y+n][x+m]:
                    temp += 1
        answer.append(temp)
sorted(answer)
print(1)
```

```python
N, M = map(int, input().split())

# 판 기본판 생성 8 x 8
Black = [['B' if (x+y)%2 == 0 else 'W' for y in range(N)] for x in range(8)]
White = [['W' if (x+y)%2 == 0 else 'B' for y in range(N)] for x in range(8)]
# 판 입력 받기
Board = [input() for i in range(N)]

# 비교 후 출력
WB_sum, BW_sum = [], []
for n in range(N-7):
  for m in range(M-7):
    WB, BW = 0, 0
    for i in range(8):
      for j in range(8):
        if Black[i][j] != Board[n+i][m+j]:
          BW +=1
        if White[i][j] != Board[n+i][m+j]:
          WB +=1
    WB_sum.append(WB)
    BW_sum.append(BW)
if min(BW_sum) >= min(WB_sum):
  print(min(WB_sum))
else:
  print(min(BW_sum))
```

## 검산:
* 속도 O(N^4), N = `50*50*50*50` = 6,250,000 1초 이내
* 메모리 : 8x8 2개 50x50 2개 = 총 5128개 -> 20512byte 넉넉.

# 설탕 배달
---



## 풀이.
---
* 시간 : 1초
* 메모리 : 128MB
* 풀이 : 브루트포스


```python
num = int(input())
answer = -1
for i in range(num//5, -1, -1):
    if i*5 + ((num - i*5)//3)*3 == num:
        answer = i + (num - i*5)//3
        break
print(answer)
```

```python
example = [[60, 50], [30, 70], [60, 30], [80, 40]]
example2 = [[10, 7], [12, 3], [8, 15], [14, 7], [5, 15]]
example3 = [[14, 4], [19, 6], [6, 16], [18, 7], [7, 11]]
def solution(sizes):
    xlist = []
    ylist = []
    for x, y in sizes:
        if x > y:
            xlist.append(x)
            ylist.append(y)
        else:
            ylist.append(x)
            xlist.append(y)
    xlist.sort(reverse=True)
    ylist.sort(reverse=True)
    return xlist[0]*ylist[0]
```

# 모의고사


```python
def solution(answers):
    len_answer = len(answers)
    answer = []
    index = 0
    per1=[1,2,3,4,5] # 5반복
    per2=[2,1,2,3,2,4,2,5] # 8반복
    per3=[3,3,1,1,2,2,4,4,5,5] # 10반복
    scores = [0,0,0]
    while len_answer:
        if per1[index%len(per1)] == answers[index]:
            scores[0] +=1
        if per2[index%len(per2)] == answers[index]:
            scores[1] +=1
        if per3[index%len(per3)] == answers[index]:
            scores[2] +=1
        index += 1
        len_answer-=1
    max_score = max(scores)
    for index, score in enumerate(scores):
        if score == max_score:
            answer.append(index+1)
    return answer
```

# 참고자료.
* iteratools.
    - `combinations([array], r)`
        * 중복 없이 r개 뽑기.
    - `combinations_with_replacement()`
        * 중복 있이 r개 뽑기.
    - `product()`
        * 중복 있이 모든 가짓수.
    - `permutations()`
        * 중복 없이 선택하는 가짓수
        * r을 통해서 선택 개수 제한/
