# 동전 0 
---



## 풀이
---
* 시간제한 1초
* 메모리제한 256MB
* 단계
    * 현재 단계에서 최선의 선택
    * 주어진 단위들에서 가장 큰 수에서 작은 순으로 나눗셈이 가능한지 여부 파악 후 나눠지면 나누기!

```python
from collections import deque
num, money = map(int, input().split())
won_list = deque()
for _ in range(num):
    won_list.appendleft(int(input()))
answer = 0
for won in won_list:
    if money//won >=1:
        answer += money//won
        money = money%won
print(answer)
```

## 검산
---
* 시간초 계산
* O(N) `10 + 10 = 20` 1초 내 가능!

# 회의실 배정
---



## 풀이
---
* 시간초 : 2초
* 메모리 : 128MB
* 최선의 선택:
    * 첫번쨰 : 회의 종료의 시간이 가장 빠른것을 선택.
    * 그 이후 : 앞의 회의종료된 시간과 가장 가까운 시간을 선택.


```python
num = int(input())
greeding_list = []
for _ in range(num):
    start, end = map(int, input().split())
    sub = end - start
    greeding_list.append([start, end, sub])
greeding_list.sort(key=lambda x:(x[1], x[0])) # 종료 시간 중 시작시간으로 정렬.
answer = 1
end_time = greeding_list[0][1]

for start, end, sub in greeding_list[1:]:

    if start < end_time: # 겹치면 종료
        pass
    else:
        answer+=1
        end_time = end
print(answer)
```

## 검산
---
* 시간초 계산
* N = 100,000
* O(N+NlogN+N = NlogN) `100,000*log(100,000) = 500,000` 2초 내 가능!

# ATM
---



## 풀이
---
* 시간초 : 1초
* 메모리 : 256MB
* 최선의 선택:
    * 가장 빠르게 끝나는 사람이 앞에서 서기


```python
num = int(input())
time_list = list(map(int, input().split()))
time_list.sort()
answer = 0
sum = 0
for time in time_list:
    sum += time
    answer += sum
print(answer)  
```

## 검산
---
* 시간초 계산
* N = 1,000
* O(N+NlogN = NlogN) `1,000*log(1,000) = 3,000` 1초 내 가능!

    *  -이후를 괄호 처리

```python
string = input()
sum = 0
temp = string.replace('+', '-')
num_list = temp.split('-')
calulater = []
answer = int(num_list[0])
index = 0
for num in num_list[:-1]:
    index += len(num)
    calulater.append(string[index])
    index+=1
index = 0
for idx in range(len(calulater)):
    cal = calulater[idx]
    if cal == '-':
        tmp1 = 0
        for i in range(idx+1, len(calulater)+1):
            tmp1 += int(num_list[i])
            index+=1
            if i == len(calulater):
                answer -= tmp1
                tmp1 = 0
            else:
                pass
        break
    elif index >= len(calulater):
        break
    elif cal == '+':
        answer += int(num_list[idx+1])
print(answer)
```

# 주유소
---
  


## 풀이
---
* 시간초 : 2초
* 메모리 : 512MB
* 최선의 선택:
    *  현재 도시 vs 이후 도시들 비교.
        * 이후 도시들 중 더 낮은비용이 있다면 낮은 도시에서 충전. 
    * 마지막 도시의 가격은 의미 X

```python
num = int(input())
length_list = list(map(int, input().split()))
money_list = list(map(int, input().split()))
money_list = money_list[:-1] # 마지막 제거
min_money = min(money_list) # 최소값 저장.
answer = 0
cost = money_list[0]
for idx in range(len(money_list)):
    money = money_list[idx]
    if money <= cost:
        cost = money
        answer += length_list[idx]*cost
    else:
        answer += length_list[idx]*cost
print(answer)
```

## 검산.
---
* n = 100,000
* O(N)
    * 시간 : 1초이내
    * 메모리 : 400MB

# 체육복
---


## 풀이
---
* 최선의 선택:
    * 처음에 본인이 잃어버리면, 본인꺼 입기
    * 그 후 빌려 줄 수 있는지 확인.

```python
n = 10
lost = [3,5,7,9]
reserve = [2,4,6,10]
def solution(n, lost, reserve):
    answer = 0
    lost_items = sorted(lost) # 정렬.
    reserve_items = sorted(reserve) # 정렬.
    temp_reserve = reserve_items.copy()
    answer = n - len(lost_items) 
    for reserve in reserve_items:
        if reserve in lost_items:
            lost_items.remove(reserve)
            temp_reserve.remove(reserve)
            answer+=1
    for lost_item in lost_items:
        if lost_item-1 in temp_reserve:
            temp_reserve.remove(lost_item-1)
            answer +=1
        elif lost_item+1 in temp_reserve:
            temp_reserve.remove(lost_item+1)
            answer+=1
    return answer
print(solution(n, lost, reserve))
```

## 검산.
----
* 시간 복잡도 O(N)
* 메모리 최대 240byte
