```python
from collections import deque
num, money = map(int, input().split())
won_list = deque()
for _ in range(num):
    won_list.appendleft(int(input()))
answer = 0
for won in won_list:
    if money//won >=1:
        answer += money//won
        money = money%won
print(answer)
        


```
