```python
from collections import deque
num, money = map(int, input().split())
won_list = deque()
for _ in range(num):
    won_list.appendleft(int(input()))
answer = 0
for won in won_list:
    if money//won >=1:
        answer += money//won
        money = money%won
print(answer)
```

```python
n = 10
lost = [3,5,7,9]
reserve = [2,4,6,10]
def solution(n, lost, reserve):
    answer = 0
    lost_items = sorted(lost) # .
    reserve_items = sorted(reserve) # .
    temp_reserve = reserve_items.copy()
    answer = n - len(lost_items) 
    for reserve in reserve_items:
        if reserve in lost_items:
            lost_items.remove(reserve)
            temp_reserve.remove(reserve)
            answer+=1
    for lost_item in lost_items:
        if lost_item-1 in temp_reserve:
            temp_reserve.remove(lost_item-1)
            answer +=1
        elif lost_item+1 in temp_reserve:
            temp_reserve.remove(lost_item+1)
            answer+=1
    return answer
print(solution(n, lost, reserve))
```
