## 알고리즘 수업 - 점근적 표기 1
---
image.png

```python
f1, f2 = map(int, input().split())
c = int(input())
n0 = int(input())

# 계산
condition = ((c-f1)*n0-f2)>=0 and (c-f1)>=0
if condition:
    print(1)
else:
    print(0)
```
