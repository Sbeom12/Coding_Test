```python
from collections import defaultdict
N, E, R = map(int, input().split())
graph = defaultdict(set)
example = [[1,4],[1,2],[2,3],[3,4]]
for ex in example:#range(E):
    node1, node2 = ex #map(int, input().split())
    graph[node1].add(node2)
    graph[node2].add(node1)
sequence = [0 for _ in range(N+1)]
visited = [False for _ in range(N+1)] # visited 배열 추가
sequence[1] = 1
count = 1
def dfs(N, R):
    global count
    visited[N] = True # 방문한 노드를 체크
    for n in graph[N]:
        if n == R or visited[n]: # 이미 방문한 노드는 건너뜀
            continue
        count += 1
        sequence[n] = count
        dfs(n, N)
dfs(1,0)

for answer in sequence[1:]:
    print(answer)
```

```python
def solution(numbers, target):
    global answer
    def dfs(idx, sum):
        global answer
        if idx == len(numbers):
            if target == sum:
                answer += 1
            return
        dfs(idx+1, sum - numbers[idx])
        dfs(idx+1, sum + numbers[idx])
    answer = 0
    dfs(0, 0)
    return answer
```

```python
solution([1, 1, 1, 1, 1], 3)
```
