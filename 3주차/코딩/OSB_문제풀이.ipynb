```python
from collections import deque
import sys
num = int(sys.stdin.readline().rstrip())
stack = deque()
for _ in range(num):
    order = list(map(int, sys.stdin.readline().rstrip().split()))
    if order[0] == 1:
        stack.append(order[1])
    elif order[0] == 2:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.pop())
    elif order[0] == 3:
        print(len(stack))
    elif order[0] == 4:
        if len(stack) == 0:
            print(1)
        else:
            print(0)
    elif order[0] ==5:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[-1])
```

# 제로
---



```python
import sys
num = int(sys.stdin.readline().rstrip())
stack = list()
total_sum = 0
for _ in range(num):
    order = int(sys.stdin.readline().rstrip())
    if order == 0:
        total_sum -= stack.pop()
    else:
        total_sum += order
        stack.append(order)
print(total_sum)
```

# 괄호

```python
import sys
num = int(sys.stdin.readline().rstrip())
def check(order):
    for word in order:
        if word == '(':
            stack.append('(')
        elif word == ')':
            if len(stack) != 0:
                stack.pop()
            else:
                return 'NO'
                
    if len(stack) == 0:
        return 'YES'
    else:
        return 'NO'
    
for _ in range(num):
    stack = list()
    order = sys.stdin.readline().rstrip()
    print(check(order))
```

# 균형잡힌 세상

```python
import sys
def check(order):
    for word in order:
        if word == '(':
            stack.append('(')
        elif word == ')':
            if '(' in stack:
                temp = 0
                for i in range(len(stack)-1, -1, -1):
                    if stack[i] == '[':
                        return 'no'
                    elif stack[i] == '(':
                        temp = i
                    break
                stack.pop(temp)
            else:
                return 'no'
        elif word == '[':
            stack.append('[')
        elif word == ']':
            if '[' in stack:
                temp = 0
                for i in range(len(stack)-1, -1, -1):
                    if stack[i] == '(':
                        return 'no'
                    elif stack[i] == '[':
                        temp = i
                    break
                stack.pop(temp)
            else:
                return 'no'
                
    if len(stack) == 0:
        return 'yes'
    else:
        return 'no'
    
while True:
    stack = list()
    order = input()
    if order == '.':
        break
    print(check(order))
```

# 도키도키 간식드리미
---



```python
num =int(input())
sequence = list(map(int, input().split()))
temp = list()
number = 1

while True:
    if len(sequence) != 0:
        if sequence[0] == number:
            sequence.pop(0)
            number+=1
        elif len(temp) != 0:
            if temp[-1] == number:
                temp.pop(-1)
                number+=1
            else:
                temp.append(sequence.pop(0))
                continue
        else:
            temp.append(sequence.pop(0))
            continue
    elif len(temp) != 0:
        if temp[-1] == number:
            temp.pop(-1)
            number+=1
    
    if len(sequence) == 0 and number != temp[-1]:
        print('Sad')
        break
    elif number == num:
        print('Nice')
        break

```

# 큐 2
---



```python
from collections import deque
import sys
num = int(sys.stdin.readline().rstrip())
que = deque()
for _ in range(num):
    order = sys.stdin.readline().rstrip().split()
    if order[0] == 'push':
        que.append(order[1])
    elif order[0] == 'pop':
        if len(que) == 0:
            print(-1)
        else:
            print(que.popleft())
    elif order[0] == 'size':
        print(len(que))
    elif order[0] == 'empty':
        if len(que) == 0:
            print(1)
        else:
            print(0)
    elif order[0] == 'front':
        if len(que) == 0:
            print(-1)
        else:
            print(que[0])
    elif order[0] == 'back':
        if len(que) == 0:
            print(-1)
        else:
            print(que[-1])    
```

# 카드2
---



```python
from collections import deque
import sys
num = int(input())
card_list = deque([v for v in range(1,num+1)])
number = 0
while len(card_list) != 1:
    if number % 2 == 0:
        card_list.popleft()
    else:
        card_list.append(card_list.popleft())
    number += 1
print(card_list[0])
```

```python
# 다른 방법.
N = int(input()) 
int_queue = [i for i in range(1, N+1)]
front = 0 # 맨앞 
end = N-1 # 맨뒤
while end>front:
  front += 1 # 맨처음 pop 효과
  int_queue.append(int_queue[front]) # 맨 뒤로 보내기
  front += 1 # 맨 뒤로 보냇으므로 그 다음으로 이동해야함.
  end += 1 # 맨 뒤가 하나 늘어나는 효과
print(int_queue[front])
```

# 요세포스 문제
---



```python
from collections import deque
import sys
n, k = map(int, input().split())
num_list = deque([v for v in range(1,n+1)])
temp = 1
Josephus = []
while len(num_list) !=0:
    if temp % k == 0:
        Josephus.append(num_list.popleft())
    else:
        num_list.append(num_list.popleft())
    temp += 1
answer = []
for i in Josephus:
    answer.append(str(i))
answer2 = ", ".join(answer)
print('<'+answer2+'>')
```

# 덱 2
---



```python
from collections import deque
import sys
num = int(sys.stdin.readline().rstrip())
stack = deque()
for _ in range(num):
    order = list(map(int, sys.stdin.readline().rstrip().split()))
    if order[0] == 1:
        stack.appendleft(order[1])
    elif order[0] == 2:
        stack.append(order[1])
    elif order[0] == 3:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.popleft())
    elif order[0] == 4:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.pop())
    elif order[0] == 5:
        print(len(stack))
    elif order[0] == 6:
        if len(stack) == 0:
            print(1)
        else:
            print(0)
    elif order[0] ==7:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[0])
    elif order[0] ==8:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[-1])
```

# 풍선 터뜨리기
---



```python
from collections import deque
num = int(input())
num_list = deque([v for v in range(1, num+1)])
command_list = list(map(int, input().split()))
answer = []
while len(num_list) !=0:
    ballon = num_list.popleft()
    answer.append(ballon)
    command = command_list[ballon-1]
    print(num_list)
    if command > 0:
        command = -(command - 1)
    else: 
        command = -command
    num_list.rotate(command)
print(*answer)
    
```
