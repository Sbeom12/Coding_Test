```python
from collections import deque
import sys
num = int(sys.stdin.readline().rstrip())
stack = deque()
for _ in range(num):
    order = list(map(int, sys.stdin.readline().rstrip().split()))
    if order[0] == 1:
        stack.append(order[1])
    elif order[0] == 2:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.pop())
    elif order[0] == 3:
        print(len(stack))
    elif order[0] == 4:
        if len(stack) == 0:
            print(1)
        else:
            print(0)
    elif order[0] ==5:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[-1])
```

```python
import sys
num = int(sys.stdin.readline().rstrip())
stack = list()
total_sum = 0
for _ in range(num):
    order = int(sys.stdin.readline().rstrip())
    if order == 0:
        total_sum -= stack.pop()
    else:
        total_sum += order
        stack.append(order)
print(total_sum)
```

```python
import sys
num = int(sys.stdin.readline().rstrip())
def check(order):
    for word in order:
        if word == '(':
            stack.append('(')
        elif word == ')':
            if len(stack) != 0:
                stack.pop()
            else:
                return 'NO'
                
    if len(stack) == 0:
        return 'YES'
    else:
        return 'NO'
    
for _ in range(num):
    stack = list()
    order = sys.stdin.readline().rstrip()
    print(check(order))
```
