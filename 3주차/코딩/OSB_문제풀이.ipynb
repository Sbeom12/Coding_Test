```python
from collections import deque
import sys
num = int(sys.stdin.readline().rstrip())
stack = deque()
for _ in range(num):
    order = list(map(int, sys.stdin.readline().rstrip().split()))
    if order[0] == 1:
        stack.append(order[1])
    elif order[0] == 2:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.pop())
    elif order[0] == 3:
        print(len(stack))
    elif order[0] == 4:
        if len(stack) == 0:
            print(1)
        else:
            print(0)
    elif order[0] ==5:
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[-1])
```

```python
from collections import deque
import sys
num = int(sys.stdin.readlines().rstrip())
que = deque()
for _ in range(num):
    order = sys.stdin.readlines().rstrip().split()
    if order[0] == 'push':
        que.append(order[1])
    elif order[0] == 'pop':
        if len(que) == 0:
            print(-1)
        else:
            print(que.popleft())
    elif order[0] == 'size':
        print(len(que))
    elif order[0] == 'empty':
        if len(que) == 0:
            print(1)
        else:
            print(0)
    elif order[0] == 'front':
        if len(que) == 0:
            print(-1)
        else:
            print(que[0])
    elif order[0] == 'back':
        if len(que) == 0:
            print(-1)
        else:
            print(que[-1])    
```
