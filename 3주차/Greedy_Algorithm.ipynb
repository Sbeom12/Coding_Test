# Greedy Algorithm(탐욕 알고리즘)
---
* 현재 상황에서 가장 좋아 보이는 것만을 선택하는 알고리즘(책)
    * 미래를 고려X
    * 각 단계에서 최선의 선택이 전체적으로 최선이길 바라는 알고리즘(약간의 기도)
    * 최종적 해답 단계에선 최적이라는 보장이 X
    * 탐욕 알고리즘로 푸는 문제들은 지역적으로 최적이면서 전역적으로 최적인 문제들.
    
* 사전에 외우고 있지 않아도 풀 수 있는 가능성이 높은 문제 유형
    * 많은 유형을 접해봐야함.
    * 창의력, 문제를 풀기 위한 최소한의 아이디어를 뗘올릴 수 있는 능력이 요구됨.
---

## 문제푸는 절차
---
* 선택 절차: 현재 상태에서 최적의 해답 선택
* 적절성 검사 : 선택된 해가 문제의 조건이 만족하는지 검사
* 해답 검사 : 원래의 문제가 해결되었는지 검사, 만약 안됬을 시 선택 절차로 돌아감.


## 탐욕 알고리즘의 조건
* 탐욕적 선택 속성 : 앞의 선택이 이후의 선택에 영향X.
* 최적 부분 구조 : 문제에 대한 최종 해결 방법은 부분 문제에 대한 최적 문제 해결 방법으로 구성.
* 위 두가지 조건이 만족하지 않으면 최적해를 구하지 못함.


## 예외.
* 매트로이드 : 어떤 특별한 구조가 있는 문제에 대해 언제나 최적해를 찾아내는 조건.
* 위 조건이 만족하지 않는 경우에도 유사 근사 알고리즘으로 문제를 풀 쑤 있다.
